### TransformDatasetを用いたデータセットの作成

Chainerではデータオーギュメンテーションをデータセットに対して適応させる便利な方法が用意されています。

`TransformDataset` を使用する事によって、データセットに対して、定義した処理を実行する事が可能です。  

詳細の情報は[公式のドキュメント](https://docs.chainer.org/en/stable/reference/generated/chainer.datasets.TransformDataset.html)を確認してください。

````python
# 必要なモジュールのインポート
from chainer.datasets import TransformDataset
from chainercv.transforms import resize
from functools import partial
````

画像の前処理の一種として画像の標準化があります。  
基本的にNNのモデルでは標準化が必須のように画像の領域でも標準化は重要な前処理です。  

まずそれぞれの画像のチャンネル毎の平均値と標準偏差を求めます。  

````python
mean = np.mean([x for x, _ in train], axis=(0, 2, 3))
std = np.std([x for x, _ in train], axis=(0, 2, 3))
````

下記の `transform` の関数でどのような前処理を加えるのかを定義しています。  
標準化の部分は平均値を引いて、標準偏差で割る処理を行なっています。  

````python
def transform(inputs, train=True):
    img, label = inputs
    img = resize(img, (224,224))
    img = img.copy()
    
    # PCA lightningの適応
    if train:
        img = transforms.pca_lighting(img, 76.5)

    # 画像の標準化
    img -= mean[:, None, None]
    img /= std[:, None, None]

    # Random flip & cropの適応
    if train:
        img = transforms.random_flip(img, x_random=True)
        img = transforms.random_crop(img, (224, 224))

    return img, label
````

````python
train = TransformDataset(train, partial(transform, train=True))
test = TransformDataset(test, partial(transform, train=False)) # フラグをFalseに返還し、必要な処理のみを適応
````

````python
chainercv.visualizations.vis_image(train[0][0])
````

````python
class VGG16(chainer.Chain):

    def __init__(self, n_out=9):
        super().__init__()
        with self.init_scope():
            self.base = L.VGG16Layers()
            self.fc = L.Linear(None, n_out)

    def __call__(self, x):
        h = self.base(x, layers=['fc7'])
        h = self.fc(h['fc7'])
        return h
````

````python
# 乱数のシード固定用の関数
def reset_seed(seed=0):
    np.random.seed(seed)
    random.seed(seed)
    if chainer.cuda.available:
        chainer.cuda.cupy.random.seed(seed)
````

````python
# CPUとGPU関連のシードをすべて固定
reset_seed(0)
````

````python
# インスタンス化
vgg16 = VGG16()
model = L.Classifier(vgg16)
````

````python
gpu_id = 0  # 使用したGPUに割り振られているID
model.to_gpu(gpu_id)
````

````python
optimizer = chainer.optimizers.Adam()
optimizer.setup(model)
````

````python
model.predictor.base.disable_update()
````

````python
batchsize = 64
train_iter = chainer.iterators.SerialIterator(train, batchsize)
test_iter = chainer.iterators.SerialIterator(test, batchsize, repeat=False, shuffle=False)
````

````python
from chainer import training
from chainer.training import extensions

epoch = 30

updater = training.StandardUpdater(train_iter, optimizer, device=gpu_id)

trainer = training.Trainer(updater, (epoch, 'epoch'), out='drive/My Drive/Colab Notebooks/result')

# バリデーション用のデータで評価
trainer.extend(extensions.Evaluator(test_iter, model, device=gpu_id))

# 学習結果の途中を表示する
trainer.extend(extensions.LogReport(trigger=(1, 'epoch'), log_name='dog-cat_dataaugumentation'))

# １エポックごとに結果をlogファイルに出力させる
trainer.extend(extensions.PrintReport(['epoch', 'iteration', 'main/accuracy', 'validation/main/accuracy', 'main/loss', 'validation/main/loss', 'elapsed_time']), trigger=(1, 'epoch'))
````

````python
trainer.run()
````
