## 画像処理の基礎

### 画像の扱い方

この章では、画像処理について学んでいきます。画像処理を学ぶにあたり、そもそも画像とは一体どんな形で扱われるのかを確認していきます。

まず、画像にはご存知の通り、高さ（Height）と幅（Width）の情報を持っています。これに加えて、奥行きを表すチャンネル（Channel）という情報も持っています。
一般的に用いられるのは、RGBという赤（Red）、緑（Green）、青（Blue）という3つのチャンネル構成になります。色の表現方法としてはCMYKカラーなども存在します。

![](images/12_01.png)

この高さと幅の単位を`pixel`で表現し、1pixel×1pixelの中にそれぞれ値が入っております。この値を**輝度**と言い、基本的には0~255の中の整数値で表現します。なぜ0~255で表現するかというと、1Byteで表現できる値が0~255の合計256個となっているからです。1Byte=8bitであり、2進数で表すと$2^8=256$となるからです。

実装する際にも確認しますが、画像データのサイズはHeight、Width、Channelの3つで表現することになります。


### フィルタとは

画像処理を理解する上で欠かせないのが**フィルタ**です。スマホの画像加工アプリ等に搭載されている機能と同じようなものになります。画像加工アプリでは、元の画像に対して、色を暗くしたり、モノクロ画像にしたり、おしゃれな画像にするといった処理を施しています。これがフィルタです。フィルタは元の画像に対して特定の処理を施すことを言います。フィルタは別名**カーネル**とも呼ばれるので覚えておいてください。また、処理前の画像のことをオリジナル画像と表現します。
以下の図は、画像処理でよく用いられるエッジ検出を行った際のイメージ図を描いております。エッジとは、簡単に言うと物体の境界線や端のようなものを意味します。
![12_03.png](images/12_03.png)




特定サイズのフィルタ（上記画像では、3×3）を用意し、左上から順番に処理を施しております。処理後、画像からエッジを検出されたものができあがっております。
これがエッジ検出用のフィルタを用いて処理が行われた際のイメージです。

それでは、実際にどのような計算が行われているのかを確認していきます。

### フィルタの計算

エッジ検出で行われているフィルタの計算は非常に簡単です。今回は、3×3のサイズで、1列目の値が-1、2列目の値が0、3列目の値が1となっているフィルタを使用します。以下の図を用いながら説明していきます。

![12_05.png](images/12_05.png)

まずは左上から始めていきます。オリジナル画像とフィルタを重ね合わせ、それぞれ対応している（重なっている）部分の数値を確認してください。それらの重なり合う数値同士をかけ算し、その計算結果の総和をエッジ検出後の左上の値として入れます。つまり、計算結果は以下のようになります。


$$
\begin{align}
\begin{array} { r } { - 1 \times 0 + 0 \times 0 + 1 \times 1 } \\ { + ( - 1 ) \times 0 + 0 \times 1 + 1 \times 2 } \\ { + ( - 1 ) \times 1 + 0 \times 3 + 1 \times 1}\\
{ = 5 }\end{array}
\end{align}
$$


計算後は、フィルタを1pixel右にずらし、再度重なり合う部分を計算、これらの操作をフィルタの右端とオリジナル画像の右端が一致するまで繰り返します。
その後、下に1pixelずらし同様の操作を行います。1pixelずらすことを**ストライド**と言い、どれくらいずらすかは基本的に1pixelとすることが多いですが、人間側で設定できるハイパーパラメータとなっております。


仮に、オリジナル画像のサイズが5×5であるとします。フィルター処理後の画像サイズはどうなるでしょうか。
答えは3×3のサイズになります。実際に5×5の画像に対し、3×3のフィルタを右端まで計算してみると理解できるのですが、3回ストライドを行った時点で右端たどり着き、それ以上ストライドできない状態となります。これは、上下左右1pixelずつ小さくした画像ができることになります。このようにそのままフィルタをかけてしまうと,
画像のサイズが小さくなってしまうため、以下のように周りの値を0で埋めてあげます。これを**padding（パディング）**と言い、この処理を施すことにより、フィルタ後の画像サイズも一定に保つことができます。


![12_06.png](images/12_06.png)


### エッジ検出とは

前節ではフィルタの計算について学びました。その際に用いたのはエッジ検出のフィルタということで、1列目が-1、2列目が0、3列目が１となっていました。
さて、この数値にはどのような意味があるのでしょうか。そもそもエッジとは具体的にどんなものなのでしょうか。

エッジは物体の境界線や端のようなものであると前述しましたが、異なった視点から考えてみます。
画像は色（輝度）という情報を扱っていますが、人間がどのようにエッジを判断しているでしょうか。色という観点から考えた際にエッジと呼ばれる部分にはどのような特徴があるでしょうか。

エッジは、**色の変化量**が大きい部分と考えられるのではないでしょうか。異なる色どうしの境界線にはエッジがあると人は認識しています。
この**変化量**が非常に重要なポイントであり、勘の良い方は数学のある言葉を思い浮かべたのではないでしょうか。
そうです、**微分**です。実は、エッジ検出には微分の考えが用いられています。




### エッジ検出の計算

エッジ検出において微分が関係していることはわかりましたが、先程のフィルタとの関係性はまだわかっておりません。数学に触れながらお伝えしていきます。

まず、以下のように横軸$x$、縦軸$f(x)$のグラフを考えます。$x$はある画像（ピクセル）の位置、$f(x)$はその$x$における輝度（色の値）をイメージしてください。

![12_07.JPG](images/12_07.png)

このある点$x$における色の変化量を求めるためには、$x$における微分（接線の傾き）を計算するのが一般的な方法ではあります。
しかし、実際に微分を行うには関数の式がわかっていないといけません。さて、色の位置と輝度の関係はどのような式で表現することができるでしょうか。
式で表すことができれば良いですが、実際には色の位置と輝度の関係を式で表現するのはあまり現実的ではありません。

そこで、ある点$x$に関しての微分は厳しいですが、着目したい点$x$から両隣の点について考えてみます。
左隣の点を$x-1$、対応する輝度の値を$f(x-1)$、右隣の点を$x+1$、対応する輝度の値を$f(x+1)$とします。

![12_08.JPG](images/12_08.png)


左右２点を通る直線というのは、$x$における接線と似たような形になります。つまり、$x$における接線の近似を表現しています。このことから、左右２点を通る直線の傾きを求めることで、近似的に$x$における接線の傾きを求めます。直線の傾きは以下のようになります。


$$
\begin{align}
\begin{aligned} a & = \frac { f ( x + 1 ) - f ( x - 1 ) } { ( x + 1 ) - ( x - 1 ) } \\ & = \frac { f ( x + 1 ) - f ( x - 1 ) } { 2 } \end{aligned}
\end{align}
$$


それでは、求めたこの傾きの値とフィルタの値の関係はどのようになっているでしょうか。ここが重要なポイントです。

フィルターの計算を少し抽象的な形で考えてみます。グラフを使用したときと同様に、着目したい点における輝度を$f(x)$、左右の値をそれぞれ$f(x-1)$、$f(x+1)$で考えます。

![12_09.JPG](images/12_09.png)


この時の計算結果は以下のようになります。

$$
\begin{align}
\begin{array} { l } { - 1 \times f \left( x-1 \right) + 0 \times f ( x ) + 1 \times f \left( x+1  \right) } { = f \left( x+1 \right) - f \left( x- 1 \right) } \end{array}
\end{align}
$$

こちらの　$f( x+1) - f( x-1 )$　という式、どこかで見覚えはないでしょうか。先程２点を通る直線の傾きを計算した際の分子の値と一致します。
分母の値は常に$2$であり、固定値となりますので、分子の値のみに注目すれば大丈夫です。

これらのことから実は、フィルタの計算は２点を通る直線の傾き、つまり微分の近似を計算していたのです。

これが、微分とエッジ検出フィルタの関係性になります。

### グレースケール変換
それでは、代表的な画像処理であるグレースケール変換を施していきます。

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

matplotlibの`imshow`では、RGBが渡される規定であるため、グレースケールでは変な色合いになってしまいます。

```python
plt.imshow(img_gray)
```

そのため、`plt.gray()`と1行加え、グレースケール画像であることを宣言してプロットを行います。

```python
plt.gray()
plt.imshow(img_gray)
```

こちらで正しく表示されました。

グレースケール変換を施したことによって、３枚あった画像が１枚になっていることがサイズを確認することでわかります。

```python
img_gray.shape
```

### フィルタをかけよう

例として紹介した横方向の微分フィルタをかけてみます。

まずは、フィルタを準備します。フィルタのことを**カーネル**とも呼び、サンプルのソースコードでは`kernel`という変数名が良く使われますので、覚えておいてください。

```python
kernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])
```

`filter2D()`を使用することで、畳み込み（convolution）の演算を行います。

```python
img_conv = cv2.filter2D(img_gray, 0, kernel)
```

こちらをプロットしてみます。

```python
plt.gray()
plt.imshow(img_conv)
```

たしかに横方向に対して、エッジが検出できていることが直感的にわかると思います。


縦方向も試してみます。

```python
kernel = np.array([
    [-1, -1, -1],
    [  0,  0,  0],
    [  1,  1,  1]
])
```

```python
img_conv = cv2.filter2D(img_gray, -1, kernel)
```

```python
plt.gray()
plt.imshow(img_conv)
```

こちらのように縦方向に色の変化が強いところを抽出することができました。

## Convolutional Neural Network（CNN）



### 従来の画像分類

CNNが出てくる前までは、どのように画像分類を行っていたでしょうか。ネコとイヌの画像を分類するという例を用いて考えていきます。
基本的には、前述したフィルタを用いて画像から分類するために必要な情報を抽出します。エッジ検出やグレースケール変換、ノイズ除去といったフィルタを使用して情報抽出していきます。これは人間の勘や今までの経験を駆使し、フィルタを選定しています。ネコを判別するにはネコの形をうまく捉えるような情報が必要だ、それならばエッジ検出を使おう、といった流れです。
しかし、画像のままではベクトル形式で情報が格納されていないため、全結合層のNNに入れてあげることができないです。そこで、ヒストグラムやHOGといった画像の特徴を表す特徴ベクトルに変換した上で、NNにかけています。

![12_10.jpg](images/12_10.png)

### CNNとは

従来の画像分類の流れは理解することができたかと思います。それでは、イヌとネコを見分けるためのフィルタにエッジ検出を用いることは本当に必要な処理なのでしょうか。もしかしたら別のフィルタの方がいいのではないでしょうか。

この問題を解決し、しっかりとゴール逆算で考えられた手法がConvoulutional Neural Netowork（CNN）です。CNNは画像の特徴を抽出するフィルタ自体をパラメータにしてしまったのです。つまり、ネコとイヌを分類するために本当に必要なフィルタを学習で求める、というゴール逆算型の発想で生まれた手法なのです。


![12_11.jpg](images/12_11.png)


### CNNの構造

それでは、CNNの構造を詳しく見ていきます。CNNは大きく分けて以下の3つの構造に分かれております。

- Convolution（畳み込み）
- Pooling（縮小）
- Fully-Connected（全結合）

順を追ってお伝えしていきます。

![12_12.png](images/12_12.png)



#### Convolution（畳み込み）

CNNの核となるのがConvolution（畳み込み）になります。これは前述したフィルタの計算を行う部分であり、情報の圧縮を行います。フィルタの値はすべてパラメータとなっており、フィルタのサイズ、ストライドの値、Paddingの有無がハイパーパラメータとなっております。Convolution後に計算されたデータは**特徴マップ**と呼ばれます。


#### Pooling（縮小）

Convolutionでチャンネル数が増え、情報も増えてしまったので、Poolingでは画像の縮小を行います。Poolingは情報を集約するようなイメージです。
Poolingにはいくつか種類があり、代表的な**MaxPooling**についてお伝えします。
　
**MaxPooling**  

基本的には前述したフィルタと同様に範囲を決めて、その範囲に対して特定の処理を施します。仮に3×3のPoolingフィルタとした際に、MaxPoolingはオリジナル画像の重なる部分に対して、**最大値**のみを抽出します。画像から最大値という特徴を取り出すわけです。下図では、範囲内の最大値である11が抽出されています。抽出後、ストライドを行いますが、MaxPoolingは基本的にフィルタのサイズと同じ分だけ動かし、計算箇所がかぶらないような設計となっております。つまり3×3のフィルタの場合は、ストライドの値が3になります。

![12_13.jpg](images/12_13.png)

最大値を抽出するMaxPoolingの他に、指定範囲の平均値を取り出す**AveragePooling**などもあります。

前述したConvolutionとPoolingを複数回行うのが一般的であり、画像分類のコンペティションで勝つようなモデルの中には、Convolution×2回、Pooling×1回を1セットとし、複数回行うものもあります。
また、Paddingにて着目すべき点はパラメータが存在しないことです。

#### Fully-Connected（全結合）

ConvolutionとPoolingを複数回行った後、最終的にできあがった複数チャンネルある画像の値を1列に並べ、NNの入力層の値として用います。
Fully-Connectedは今までのNNの構造であり、中間層を挟んで、出力層までの流れとなっております。
