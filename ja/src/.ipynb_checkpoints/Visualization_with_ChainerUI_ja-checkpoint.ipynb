# ChainerUIで学習経過を可視化

ChainerUIはChainer上の実験中の学習ログの可視化や実験管理機能を追加するパッケージです。  
ChainerUIの情報は[公式ドキュメント](https://github.com/chainer/chainerui)に記載がありますので、詳細はこちらから確認してください。  

ChainerUIの主な機能には下記の3点が挙げられます。  

**ChainerUIの主な機能**

- 学習ログの可視化
- 実験履歴閲覧
- 実行中の実験に対する操作 (スナップショット・学習ハイパーパラメーター修正)

これらの機能によって学習中の実験ジョブの進捗確認や、複数の実験ジョブを手軽に管理することが可能になります。

````python
# 必要なモジュールのインポート
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

### モデルの構築

今回はシンプルなネットワークで学習を行います。

````python
import chainer
import chainer.functions as F
import chainer.links as L
````

````python
class CNN(chainer.Chain):
    
    def __init__(self, n_mids, n_out=10):
        super().__init__()
        with self.init_scope():
            self.conv1 = L.Convolution2D(1, 3, ksize=3, pad=1)
            self.fc1 = L.Linear(None, n_mids)
            self.fc2 = L.Linear(None, n_out)
            
    def __call__(self, x):
        h = self.conv1(x)
        h = self.fc1(h)
        return self.fc2(h)
````

````python
# 再現性の確保
np.random.seed(0)
````

````python
# モデルの宣言
model = L.Classifier(CNN(100, 10))
````

````python
#Optimizerの定義
optimizer = chainer.optimizers.MomentumSGD()
optimizer.setup(model)
````

````python
#iteratorの定義
batchsize = 128
train_iter = chainer.iterators.SerialIterator(train, batchsize)
valid_iter = chainer.iterators.SerialIterator(valid, batchsize, repeat=False, shuffle=False)
````

````python
#Updaterの定義
epoch = 20
updater = chainer.training.StandardUpdater(train_iter, optimizer, device=-1)
````

````python
# Trainerの定義
trainer = chainer.training.Trainer(updater, (epoch, 'epoch'), out='result/result1')
````

````python
# Extensionsの定義
from chainer.training import extensions

trainer.extend(extensions.Evaluator(valid_iter, model, device=-1))
trainer.extend(extensions.LogReport())
trainer.extend(extensions.observe_lr())
trainer.extend(extensions.PrintReport(['epoch',  'elapsed_time', 'main/loss', 'validation/main/loss','main/accuracy', 'validation/main/accuracy', 'lr']))
````

````python
trainer.run()
````

### 引数の値を表示

引数の値を表示するためには引数を辞書型に格納して、保存する必要があります。  
引数を辞書型に格納するには `easydict` というライブラリを使用すると便利です。  
※スクリプトでコードを書く際の引数の定義はこちらの[GitHub](https://github.com/chainer/chainerui/blob/master/examples/train_mnist.py)のページを確認してください。  

````python
# easydictのインストール
!pip install easydict
````

````python
# 引数の定義
import easydict
args = easydict.EasyDict({
    "batchsize":128,
    "epoch":20,
    "frequency":1,
    "resume":'',
    "gpu":-1,
    "out":'result/result2',
    "unit":50
})
````

この定義した引数は `from chainerui.utils import save_args` で保存することが可能です。   
`save_args(引数を格納した変数, 保存先)` と指定すれば保存が実行されます。   

辞書型で格納された引数は `args.Key名` で呼び出すことができます。  

````python
from chainerui.utils import save_args
save_args(args, args.out)
````

### 複数の学習結果を表示

複数の学習結果を表示するためにはプロジェクトの階層の下に複数のフォルダを作成し、そのフォルダ内にそれぞれ学習結果を保存した`log`という名前のファイルを用意すれば表示することができます。  
先ほど定義した引数`args.out`には現在`'result/result2'`という値が入っており、  
プロジェクトが存在する`result`の階層の下に`result2`があり、その中に`log`ファイルが保存されるように指定しています。  


```bash
作業フォルダ ─ mnist.ipynb
     　　　　　　　　　　　　　 └─  result
                       └ mnist-project   
                       └ result1 ─ log 
                       └ result2 ─ log 

```

````python
# Trainerの定義
trainer = chainer.training.Trainer(updater, (args.epoch, 'epoch'), out=args.out)
````

### 学習中のsnapshotの作成&学習中の学習係数の変更

学習中のsnapshotの作成と学習中の学習係数の変更行うには`Extensions`の部分で1つ設定を加えるだけで実現することができます。  

`from chainerui.extensions import CommandsExtension`の`trainer.extend(CommandsExtension())`を記述して完了です。

````python
# Extensionsの設定
from chainer.training import extensions
from chainerui.extensions import CommandsExtension

trainer.extend(extensions.Evaluator(valid_iter, model, device=args.gpu))
trainer.extend(extensions.LogReport())
trainer.extend(extensions.observe_lr())
trainer.extend(extensions.PrintReport(['epoch',  'elapsed_time', 'main/loss', 'validation/main/loss','main/accuracy', 'validation/main/accuracy', 'lr']))
trainer.extend(CommandsExtension())
````

````python
trainer.run()
````
